```python

data = pd.read_csv('/kaggle/input/spam-email-classification/email.csv')
data.columns
```

```python
input_data=data['Message']
output_data=data['Category']
```

```python
vectorizer=TfidfVectorizer(stop_words='english',max_features=5000)
conversion=vectorizer.fit_transform(input_data)
```

```python
X_train , X_test, Y_train , Y_test =train_test_split(conversion , output_data , test_size=0.2 , random_state=42)
```

```python
model=LogisticRegression()
model.fit(X_train , Y_train)
```

```python
tested=model.predict(X_test)
```

```python
print(f"Accuracy score of the model is: {accuracy_score(tested , Y_test)}")
```

# Now we will** test** our model
